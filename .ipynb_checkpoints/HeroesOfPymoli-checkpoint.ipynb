### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(5)
```

## Player Count

* Display the total number of players


```python
#Calculate total number of players
total_players = purchase_data["SN"].unique()
total_players = len(total_players)
print(f'Total Players: {total_players}')
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate Total number of unique items
uniqueItemIds = purchase_data["Item ID"].nunique()

# Calculate Total Number of Purchases
totalNumberPurchases = purchase_data["Purchase ID"].count()

# Calculate total revenue
totalRevenue = purchase_data["Price"].sum()

# Calculate average purchase price
averagePrice = round(totalRevenue/totalNumberPurchases, 2)

```

```python
# Create and print a 'Summary Table' data frame 
summary_table = pd.DataFrame({"Number of Unique Items":uniqueItemIds,
                              "Average Price": [averagePrice],#.map("${:,.2f}".format),
                              "Number of Purchases":totalNumberPurchases,
                              "Total Revenue": f'${totalRevenue}',
                              })
summary_table
    
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicate records for each player to get a subset with unique players only
unique_names_subset = purchase_data.drop_duplicates(subset = "SN", keep = "first")

# Count number of players in each gender category
genderCount = unique_names_subset.loc[:, "Gender"]
genderCount = genderCount.value_counts()

# Calculate percent of players in each gender category
percent = [round(name/total_players *100, 2) for name in genderCount]
```

```python
# Create and print a 'Summary Table' data frame 
summary_table = pd.DataFrame({"Total Count":genderCount,
                              "Percentage of Players":percent
                            })
summary_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group purchase_data by gender and calculate number of purchases
purchaseCount = purchase_data.groupby("Gender").count()["Price"]

# Calculate Average Purchase Price by gender
avgPurchasePrice = purchase_data.groupby("Gender").mean()["Price"]

# Calculate Total Price by gender
totalPurchase = purchase_data.groupby("Gender").sum()["Price"]

# Calculate Averave Total Purchase per Person by gender
AvgTotalPurchasePerson = round(totalPurchase/genderCount, 2)

# Create, format and print a 'Purchasing Analysis' data frame 
PurchasingAnalysis = pd.DataFrame({
    "Purchase Count": purchaseCount,
    "Average Purchase Price": avgPurchasePrice.astype(float).map("${:,.2f}".format),
    "Total Purchase Value":totalPurchase.astype(float).map("${:,.2f}".format),
    "Average Total Purchase per Person":AvgTotalPurchasePerson.astype(float).map("${:,.2f}".format)
})
PurchasingAnalysis


# .astype(float).map("${:,.2f}".format)
# formatedPurchasingAnalysis = PurchasingAnalysis.copy()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages and corresponding labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
labels = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins
unique_names_subset["Age Group"]= pd.cut(unique_names_subset["Age"], bins, labels=labels)

#Calculate total count per Age group
groupCount = unique_names_subset.groupby("Age Group").count()["Price"]

#Calculate percent of players per Age group
percentPlayers = [round(value/total_players * 100,2) for value in groupCount]
```

```python
# Create and print a 'Age Demographics' data frame 
ageDemographics = pd.DataFrame({"Total Count":groupCount,
                                "Percentage of Players":percentPlayers})
ageDemographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data data frame by age
purchase_data["Age Group"]= pd.cut(purchase_data["Age"], bins, labels=labels)

# Calculate Purchase Count by age
age_purchaseCount = purchase_data.groupby("Age Group").count()["Price"]

# Calculate Average Purchase Price by age
age_avgPurchasePrice = purchase_data.groupby("Age Group").mean()["Price"]

# Calculate Total Purchase Value by age
age_totalPurchase = purchase_data.groupby("Age Group").sum()["Price"]

# Calculate Average Total Purchase per Person by age
age_AvgTotalPurchasePerson = age_totalPurchase/groupCount

# Create, format and print a 'Age Purchasing Analysis' data frame 

agePurchasingAnalysis = pd.DataFrame({
    "Purchase Count":age_purchaseCount,
    "Average Purchase Price": age_avgPurchasePrice.astype(float).map("${:,.2f}".format),
    "Total Purchase Value":age_totalPurchase.astype(float).map("${:,.2f}".format),
    "Average Total Purchase per Person":age_AvgTotalPurchasePerson.astype(float).map("${:,.2f}".format)
})
agePurchasingAnalysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate Total Purchase Value
totalPurchaseValue = purchase_data.groupby("SN").sum()["Price"]

# Calculate Purchase Count by each player
purchaseCount= purchase_data.groupby("SN").count()["Price"]

# Calculate Average Purchase Price
averagePurchasePrice= purchase_data.groupby("SN").mean()["Price"]

# Create, format and print a 'Top Spender' data frame
topSpender = pd.DataFrame({
    "Purchase Count": purchaseCount,
    "Average Purchase Price": averagePurchasePrice.astype(float).map("${:,.2f}".format),
    "Total Purchase Value": totalPurchaseValue
})

#Sort the Total Purchase Value column in descending order
sortedTopSpender = topSpender.sort_values(["Total Purchase Value"], ascending=False)

# Format the "Total Purchase Value" column in the sorted dataset 
sortedTopSpender["Total Purchase Value"] = sortedTopSpender["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display the 'Sorted Top Spender' dataframe
sortedTopSpender.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
popularItemsDf = purchase_data[["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name. 
grouped_popularItemsDf = popularItemsDf.groupby(["Item ID","Item Name"])

# Calculate purchase count by Item name and item Id
popularItem_purchaseCount = grouped_popularItemsDf.count()["Price"]

# Calculate Total Purchase Value per item
popularItem_totalPurchaseValue = grouped_popularItemsDf.sum()["Price"]

# Calculate Item Price
itemPrice = popularItem_totalPurchaseValue/popularItem_purchaseCount

# Create, format and print a 'Popular Items' data frame 
popularItems = pd.DataFrame({
    "Purchase Count": popularItem_purchaseCount,
    "Item Price": itemPrice,
    "Total Purchase Value": popularItem_totalPurchaseValue
})
# Sort the purchase count column in descending order
sortedPopularItems = popularItems.sort_values(["Purchase Count"], ascending=False)

# Format the "Item Price" and "Total Purchase Value" column in the sorted data frame
sortedPopularItems["Item Price"] = sortedPopularItems["Item Price"].astype(float).map("${:,.2f}".format)
sortedPopularItems["Total Purchase Value"] = sortedPopularItems["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display a preview of the data frame
sortedPopularItems.head(5)

```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort the above table by total purchase value in descending order
mostProfitableItems = popularItems.sort_values(["Total Purchase Value"], ascending=False)

# Format the "Item Price" and "Total Purchase Value" column in the sorted data frame
mostProfitableItems["Item Price"] = mostProfitableItems["Item Price"].astype(float).map("${:,.2f}".format)
mostProfitableItems["Total Purchase Value"] = mostProfitableItems["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# Display a preview of the data frame
mostProfitableItems.head(5)
```
